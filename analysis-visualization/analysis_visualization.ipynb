```python
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.linear_model import LinearRegression
dia = need_data.loc[:,'view'].values
price = need_data.loc[:,'like'].values
print(dia)
print(price)
```

```python
model = LinearRegression()  # 创建模型
X = dia.reshape((-1,1))
y = price
model.fit(X, y)  # 拟合
dia_x = []
for d in dia:
    dia_x.append([d])
X2 = dia_x # 取两个预测值
y2 = model.predict(X2)  # 进行预测
print(y2)  # 查看预测值
```

```python
plt.figure(figsize=[15, 10])
plt.xlabel("播放量")
plt.ylabel("点赞量")
plt.plot(dia, price, 'k.')
plt.plot(X2, y2, 'g-')  # 画出拟合曲线
plt.show()
```

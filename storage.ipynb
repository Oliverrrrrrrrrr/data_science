```python
import csv
csvf = open('bili.csv','r',encoding="utf-8")
reader = csv.reader(csvf)
BInfo = []
for row in reader:
    BInfo.append(row)
csvf.close()
print(BInfo[1])
```

```python
sql = 'INSERT INTO `bili`.`biliweek` (`week`, `bvid`, `category`, `coin`, `favorite`, `like`, `mid`,  `name`, `playlink`, `reply`,`title`, `view`) VALUES (%s, %s,%s, %s, %s, %s, %s,%s, %s, %s, %s, %s)'
db = pymysql.connect(host='localhost',user='root',\
                     password='ouyangyi2002',port=3306)
cursor = db.cursor()
for i in range(1,len(BInfo)):
    try:
        cursor.execute(sql, BInfo[i])
        db.commit()
        print("insert successfully")
    except Exception as e:
        db.rollback()
        print("errors:",e)
db.close()
```

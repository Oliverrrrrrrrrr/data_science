```python
import pandas as pd
import matplotlib
data = pd.read_csv("bili.csv") 
data
```

```python
need_columns = ['aweek','category', 'title', 'name','view',  'reply','favorite', 'coin', 'like']
need_data = pd.DataFrame(data, columns=need_columns)
need_data.clip()
```

```python
#点赞量与播放量关系
pd.DataFrame(need_data, columns=["view", "like"]).plot.scatter(x=["view"], y=["like"])
```

```python
#硬币与播放量关系
pd.DataFrame(need_data, columns=["view", "coin"]).plot.scatter(x=["view"], y=["coin"])
```

```python
#收藏与播放量关系
pd.DataFrame(need_data, columns=["view", "favorite"]).plot.scatter(x=["view"], y=["favorite"])
```

```python
#评论与播放量关系
pd.DataFrame(need_data, columns=["view", "reply"]).plot.scatter(x=["view"], y=["reply"])
```

```python
#提取title
from collections import Counter # 用作词频统计
import jieba
from pyecharts import options as opts
from pyecharts.charts import WordCloud
titles = need_data.loc[:, "title"]
titles
```

```python
# 导入jieba中的的tag抽取包
from jieba import analyse
word_data = []
for title in titles:
    word_list = analyse.extract_tags(title)
    word_data += word_list
new_word_counter = Counter(word_data)
```

```python
#生成词云
from wordcloud import WordCloud
import matplotlib.pyplot as plt
font = r'C:\Windows\Fonts\STCAIYUN.TTF'

my_wordcloud = WordCloud(scale=4,font_path=font,background_color='white',
                         max_words = 100,max_font_size = 60,random_state=40).generate(" ".join(word_data))


#显示生成的词云
plt.figure(figsize=[15, 10])
plt.imshow(my_wordcloud)
plt.axis("off")
plt.show()
```
